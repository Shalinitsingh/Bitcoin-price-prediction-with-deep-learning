```python
**Exploratory Data Analysis**

import pandas as pd #for dataframe
import numpy as np #for variety of mathematical operations on arrays
import seaborn as sns #for  making statistical graphics
import matplotlib.pyplot as plt #for plotting

Mounting Google drive

from google.colab import drive
drive.mount('/content/drive')

Reading data from CSV file

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project/Raw.csv')
df

Checking the information of dataframe

df.info()

df['Date'] = pd.to_datetime(df['Date']) #changing the datatype of date
df.info()

Checking for null values

df.isnull().sum()

Removing null values

df = df.dropna()
df

Checking the data type

df.dtypes

**Univariate Data Visualizations**

Plotting bar plot 

a = sns.barplot(x ='Date', y = 'Price', data = df)
plt.xticks([])
plt.show()

a = sns.barplot(x ='Date', y = 'Transaction value', data = df)
plt.xticks([])
plt.show()

Plotting scattered plot

plt.scatter(df.index, df['Price'], color = 'turquoise')
plt.ylabel('Price (in USD)')
plt.xlabel('Date')
plt.show()

plt.scatter(df.index, df['Transaction value'], color = 'turquoise')
plt.ylabel('Transaction value')
plt.xlabel('Date')
plt.show()

Plotting line plot

plt.figure(figsize = (15, 5))
ax = df.plot('Date', 'Price', color='green', linewidth=0.8)
ax.set_ylabel('Price (in USD)')
ax.set_xlabel('Date')
ax.tick_params(axis='x', rotation = 70,)
ax.set_xticklabels(df['Date'], fontsize = 8)


plt.figure(figsize = (15, 5))
ax = df.plot('Date', 'Transaction value', color='green', linewidth=0.8)
ax.set_ylabel('Transaction value')
ax.set_xlabel('Date')
ax.tick_params(axis='x', rotation = 70,)
ax.set_xticklabels(df['Date'], fontsize = 8)

Plotting density plot

plt.figure(figsize = (5, 5))
df['Transaction value'].plot(kind = 'density')

**Bivariate Data Visualizations**

a = df.iloc[0]
a

plt.figure()
ax = df.plot('Date', 'Transactions', color='green', linewidth=0.8)
ax.set_ylabel('Transactions', color='green')
ax.tick_params(axis='y', labelcolor='green')
ax2 = df.plot('Date', 'Price', color='red', linewidth=0.5, secondary_y=True, ax=ax)
ax2.set_ylabel('Price', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.figure()
ax = df.plot('Date', 'Size', color='green', linewidth=0.8)
ax.set_ylabel('Size', color='green')
ax.tick_params(axis='y', labelcolor='green')
ax2 = df.plot('Date', 'Sent by address', color='orange', linewidth=0.5, secondary_y=True, ax=ax)
ax2.set_ylabel('Sent by address', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.figure()
ax = df.plot('Date', 'Difficulty', color='green', linewidth=0.8)
ax.set_ylabel('Difficulty', color='green')
ax.tick_params(axis='y', labelcolor='green')
ax2 = df.plot('Date', 'Hashrate', color='blue', linewidth=0.5, secondary_y=True, ax=ax)
ax2.set_ylabel('Hashrate', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.figure()
ax = df.plot('Date', 'Mining profitability', color='green', linewidth=0.8)
ax.set_ylabel('Mining profitability', color='green')
ax.tick_params(axis='y', labelcolor='green')
ax2 = df.plot('Date', 'Transaction fees', color='cyan', linewidth=0.5, secondary_y=True, ax=ax)
ax2.set_ylabel('Transaction fees', color='cyan')
ax2.tick_params(axis='y', labelcolor='cyan')

plt.figure()
ax = df.plot('Date', 'Confirmation time', color='green', linewidth=0.8)
ax.set_ylabel('Confirmation time', color='green')
ax.tick_params(axis='y', labelcolor='green')
ax2 = df.plot('Date', 'Transaction value', color='turquoise', linewidth=0.5, secondary_y=True, ax=ax)
ax2.set_ylabel('Transaction value', color='turquoise')
ax2.tick_params(axis='y', labelcolor='turquoise')


**Multivariate Data Visualizations**

Plotting Matrix Scatterplot

pd.plotting.scatter_matrix(df.loc[:, "Transactions":], diagonal="kde",figsize=(40,35), color='green')
plt.tight_layout()
plt.show()

Profile plot 

ax = df.loc[:, "Transactions":].plot(figsize=(20,15))
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```
